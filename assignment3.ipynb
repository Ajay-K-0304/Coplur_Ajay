```python
#Dictionary
dct={0:'Ajay',1:'Kumar',2:'Kumawat','Branch':'IT','Height':182,'Weight':60}
#Accessing Keys and their values
print(dct[0],dct[1],dct[2])
#All keys and values
print(dct.keys(),dct.values())
#Finding particular key's value
print(dct.get('Branch'))
dct.pop(1)
copiedDct=dct.copy()
print(copiedDct)
copiedDct.clear()
print(copiedDct)
```

```python
#Tuple
tup1=tuple("SKITJAIPUR")
print(tup1)
#accessing tuples
print(tup1[0])
print(tup1[0:4])
#Unpacking
tup2=(23,45,67)
a,b,c=tup2
print(a,b,c)
#Concatenation
print(tup1+tup2)
freqI=tup1.count('I')
print("Freq of I:",freqI)
```

```python
#Set
st={"Apple","Banana","Mango",2,3,4,40.5,2,"Banana"}
print(st)
#Accessing using loop
for x in st:
    print(x,end=',')
print()
#Adding items
st.add('Papaya')
lt=[10,20,30]
st.update(lt)
print(st)
#Removing items
st.remove('Banana')
st.pop()
st.discard(4)
print(st)
#UNION INTERSECION DIFFERENCE
set1={'a','b','COMMON','c'}
set2={1,2,3,'COMMON'}
set3=set1.union(set2)
print("UNION:",set3)
set4=set1.intersection(set3)
print("INTERSECTION:",set4)
set5=set1.difference(set2)
print("DIFFERENCE:",set5)
```

```python
#Mathematical operations on two numbers
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
def maths(num1,num2):
    print("Sum:",num1+num2)
    print("Sub:",num1-num2)
    print("Mul:",num1*num2)
    print("Div:",num1/num2)
maths(num1,num2)
```

```python
#Palindrome checker
num=int(input("Enter a number:"))
org=num
reversed=0
while num!=0:
    digit=num % 10
    reversed = reversed * 10 + digit
    num //= 10
if org==reversed:
    print(f"{org} is palindrome")
else:
    print(f"{org} is not palindrome")
```
