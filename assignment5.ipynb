```python
#Creating a csv file and writing into it
import csv
data=[]
data.append(["Name","Adress","Mobile","Email"])
for i in range(1,4):
    inp=input("Enter Name,Address,Mobile,Email:")
    inp=inp.split()
    data.append(inp)
with open('data.csv','w',newline='') as file:
    writer=csv.writer(file)
    for row in data:
        writer.writerow(row)
```

```python
#Creating database
import sqlite3
db=sqlite3.connect("practice.db")
```

```python
#Creating tables
db.execute("Create table student(rollNo Integer Primary Key Autoincrement,name varchar(40),course varchar(30))")
db.execute("Create table courses(cid Integer Primary Key,cName varchar(40),duration integer not null)")
db.commit()
```

```python
db.execute("Insert into student(name,course) values ('Ajay','Python'),('Aditya','Python'),('Aishvarya','Java')")
db.commit()
```

```python
db.execute("Insert into courses values(1,'Python',6),(2,'Java',5),(3,'C++',2)")
db.commit()
```

```python
#Performing select
db.execute("Create view stud_cour as Select s.rollNo,s.name,s.course,c.cid from student s join courses c on s.course=c.cName")
```

```python
data=db.execute("select * from stud_cour")
rows=data.fetchall()
print(rows)
```
