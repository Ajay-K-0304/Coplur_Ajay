```python
s_name=input("Enter student's name:")
s_class=input("Enter student's class:")
m1=int(input("Enter marks for sub1:"))
m2=int(input("Enter marks for sub2:"))
m3=int(input("Enter marks for sub3:"))
m4=int(input("Enter marks for sub4:"))
m5=int(input("Enter marks for sub5:"))
print("Result".center(20,'-'))
total_marks=m1+m2+m3+m4+m5
perct=(total_marks/500)*100
print("Student\tClass\tPercentage")
print(s_name,s_class,perct,sep=" ")
```

```python
s1="London"
s2="is capital of England"
res=s1+" "+s2
print("Concatenated:",res)
print(res.lower(),res.upper(),res.swapcase(),sep="\t")
print(res.title(),res.capitalize(),res.casefold(),sep="\t")
print(res.center(40,'-'))
print("Index of char 'c' in res:",res.find('c'))
print("Freq of char 'c' in res:",res.count('c'))
print("Result ending with 'd':",res.endswith('d'))
print("Index of substring 'England':",res.find('England'))
print("Replacing England with India:",res.replace("England","India"))
print("Is numeric:",res.isnumeric())
print("Is digits:",res.isdigit())
print("Is alphanumeric:",res.isalnum())
print("Is spaced string:",res.isspace())
```
