```python
import pandas as pd
df1=pd.DataFrame({'Name':['A1','B1','C1','D1'],
                  'RollNo':[10,12,13,14],
                  'Class':['G1','G2','G3','G4']
                  })
df2=pd.DataFrame({'Subject':['English','Hindi','Maths','Bio'],
                  'RollNo':[10,12,13,14]})
print(df1)
print(df2)
```

```python
#Inner merge
res=df1.merge(df2,how='inner',on='RollNo')
print(res)
```

```python
#Left merge and right merge joins
df2=pd.DataFrame({'Subject':['S1','S2','S3','S4','S5'],
                  'RollNo':[10,12,13,14,15]})
r1=df1.merge(df2,on='RollNo',how='right')
print(r1)
df1=pd.DataFrame({'Name':['A1','B1','C1','D1'],
                  'RollNo':[10,12,13,14],
                  'Class':['G1','G2','G3','G4']
                  },index=[1,2,3,4])
df2=pd.DataFrame({'Subject':['English','Hindi','Maths','Bio','SST'],
                  'RollNo':[10,12,13,14,15]
                  },index=[1,2,3,4,5])
r2=df1.merge(df2)
print(r2)
```

```python
#Join
print("Joining by one column")
rs3=df1.join(df2,lsuffix='X',rsuffix='Y')
print(rs3)
df3=pd.DataFrame({'Subject':['S1','S2','S3','S4'],
                  'Class':['G1','G2','G3','G4'],
                  'Marks':[100,90,92,89]})
print("Joining by multiple keys")
rs4=df1.join(df2,lsuffix='First',rsuffix='Right')
print(rs4)
```

```python
#Merging and concating three dataframes
df1=pd.DataFrame({1:[1,2,3,4],
                  2:['A','B','C','D'],
                  3:['V1','V2','V3','V4']})
df2=pd.DataFrame({4:[9,8,7,6],
                  5:['E','F','G','H'],
                  6:['B1','B2','B3','B4']})
df3=pd.DataFrame({1:[1,2,3,4],
                  2:['I','J','K','L'],
                  3:['V1','V2','V3','V4']})
#Concating first two
cc=pd.concat([df1,df2],axis=1)
print(cc)
#Merging with third dataframe
fs=cc.merge(df3,on=[3],how='left')
print(fs)
```

```python
# | Feature      | `merge()`                    | `join()`              |
# | ------------ | ---------------------------- | --------------------- |
# | Similar to   | SQL JOIN                     | Index-based join      |
# | Works on     | Columns                      | Mostly Index          |
# | Flexibility  | More flexible                | Simpler               |
# | Common usage | Combining using keys/columns | Combining using index |
#Merge work on common column 
df1 = pd.DataFrame({
    'id':[1,2,3],
    'name':['Ajay','Rahul','Aman']
})
df2 = pd.DataFrame({
    'id':[1,2,3],
    'marks':[90,85,88]
})
result = pd.merge(df1, df2, on='id')
print(result)
#Join works on common indexes
df1 = pd.DataFrame({
    'name':['Ajay','Rahul','Aman']
}, index=[1,2,3])
df2 = pd.DataFrame({
    'marks':[90,85,88]
}, index=[1,2,3])
result = df1.join(df2)
print(result)
```
