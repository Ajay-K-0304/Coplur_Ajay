```python
#Printing multiple objects
ls=[1,2,3,4]
dt={1:'Ajay',2:'Kumar',3:'Kumawat'}
tp=(3,4,5,2,1)
st={1,1,4,5,3}
print("List:",ls)
print("Dictionary:",dt)
print("Tuple:",tp)
print("Set:",st)
a=10
b=12
c=23
d=34
print("Using separator:")
print(a,b,c,d,sep=",")
print("Using end:")
print(a,end='-')
print(b,end='-')
print(c,end='-')
print(d,end='')
```
