```python
#Max of 3 numbers
def maxOf3(n1,n2,n3):
    if n1>n2:
        if n1>n3:
            print("Max:",n1)
        else:
            print("Max:",n3)
    else:
        if n2>n3:
            print("Max:",n2)
        else:
            print("Max:",n3)
maxOf3(2,10,4)
```

```python
#Distinct elements of a list
def distinct(ls):
    oupL=[]
    st=set()
    for x in ls:
        st.add(x)
    for y in st:
        oupL.append(y)
    return oupL
inpL=[2,3,4,2,3,5,6]
ans=distinct(inpL)
print("Distinct list:",ans)
```

```python
#Multiply all numebrs in list
def multList(lst):
    mul=1
    for x in lst:
        mul*=x
    print("Multiplication of list items:",mul)
inp=[1,2,3,4,5,6,7,8]
multList(inp)
```

```python
#Factorial
def factorial(num):
    if num==0 or num==1:
        return 1
    return num * factorial(num-1)
print("Factorial:",factorial(5))
```

```python
#Reverse string
def revStr(str):
    ans=""
    for x in str[::-1]:
        ans+=x
    return ans
ans=revStr(str="HELLO")
print("Reversed:",ans)
```

```python
#Range checker
def rangeCheck(x,min,max):
    if x>=min and x<=max:
        return True
    else:
        return False
if rangeCheck(10,2,20):
    print("10 lies in range")
else:
    print("10 does not lie in range")
```

```python
#Even nubers in a list
def findEven(lst):
    for x in lst:
        if x%2==0:
            print(x)
lst=[1,2,3,4,5,6,7,8]
findEven(lst=lst)
```

```python
#Prime checker
def isPrime(n):
    if n <= 1:
        return False
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True
n=9
if isPrime(9):
    print(f"{n} is Prime")
else:
    print(f"{n} is composite")
```

```python
#String uppercase lowercase letters
def caseCounter(inp):
    upCase=0
    lwCase=0
    for x in inp:
        if x.isupper():
            upCase+=1
        else:
            lwCase+=1
    print(f"Uppercase letter:{upCase} and Lowercase letter:{lwCase}")
caseCounter("ElePhAnt")
```
